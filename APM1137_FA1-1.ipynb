> To begin, $g(x)$ is said to be continious because it is an exponential function of $x$ with base of $1/2 = 2^{-1}$
>
> As $x$ approaches to $\infty$, $g(x)$ approaches to $0$, hence it is a decreasing function
> 
> Consider the interval $x \in [\frac{1}{3}, 1]$, hence let $a = \frac{1}{3}$ and $b = 1$
>
> Determine $g(a)$ and $g(b)$ as follows:
> > $g(a) = g(\frac{1}{3}) = 2^{-\frac{1}{3}} \approx 0.7937$
> > 
> > $g(b) = g(1) = 2^{-1} = 0.5000$
> >
> > Remember that, $g(x)$ is a decreasing function as x increases, thus this shows that $a < g(b) < g(a) < b$ hence $g(x) \in [a,b]$ whenever $x \in [a, b]$
>
> Then $g$ has at least one fixed point in $[a,b]$
>
> Additionally, the derivative of $g(x)$ is as follows:
> > $g'(x) = -\frac{\ln(2)}{2^x} = -\ln(2) * 2^{-x} = \ln(1/2) * 2^{-x} = \ln(1/2) * g(x) \approx -0.6931 * g(x)$
> >
> > Moreover, $|g'(x)| \approx |(-0.6931) * g(x)| = 0.6931 * |g(x)|$
> >
> > Evaluate $|g'(a)| \approx 0.6931 * |g(a)| = 0.6931 * 2^{-\frac{1}{3}} \approx 0.550151281795 < 1 = b$
> >
> > Also, $|g'(b)| \approx 0.6931 * |g(b)| = 0.6931 * 2^{-1} \approx 0.34657359028 > \frac{1}{3} = a$
> >
> > Given $|g'(b)| < |g'(a)|$ and becuase $|g'(a)| < b$ and $a < |g'(b)|$
> >
> > Hence $a < |g'(b)| < |g'(a)| < b$ for $x \in [a,b]$, therefore $|g'(x)| \in (a,b)$
> > 
> Therefore, $g$ indeed satisfies all the hypothesis of the theorem and has a unique fixed point on $[-1, 1]$

<h4>Determine the interval $[a,b]$ on which fixed point iteration will converge for $x = \frac{5}{x^2} + 2$. Estimate the number of iterations neccesary to obtain approximations accurate to within $10^{-5}$</h4>

> We will use the fixed-point theorem's convergence condition to determine the interval $[a,b]$ provided $g(x) = \frac{5}{x^2} + 2$, hence we require $|g'(x)| < 1$
> > Evaluate $|g'(x)|= |\frac{d}{dx}(\frac{5}{x^2} + 2)|=|\frac{-10}{x^3}| < 1$
> > 
> > Simplify $|\frac{-10}{x^3}| = \frac{|-10|}{|x^3|} = \frac{10}{|x^3|} < 1$
> >
> > Hence $|x^3| > 10$ or $|x| > \sqrt[3]{10} \approx 2.154$
> >
> > This means, either $2.154 < x$ or $x < -2.154$ will be our feasible interval choices as $x \in [a,b]$
> >
> > We will take $x\in [2.154, \infty)$ as the answer to number 6

````python
# Define the function g(x)
def g(x):
    return 5 / x**2 + 2

# Define the derivative g'(x)
def g_prime(x):
    return -10 / x**3

# Fixed-point iteration method
def fixed_point_iteration(g, x0, tolerance, max_iterations=1000):
    x_n = x0
    for n in range(max_iterations):
        x_next = g(x_n)
        if abs(x_next - x_n) < tolerance:
            return x_next, n + 1
        x_n = x_next
    return x_n, max_iterations

# Initial guess and tolerance
x0 = 2.154  # Initial guess greater than 2.154
tolerance = 1e-5

# Run the fixed-point iteration
result, iterations = fixed_point_iteration(g, x0, tolerance)

# Output the result and number of iterations
print(f"Fixed point: {result}")
print(f"Number of iterations: {iterations}")
````

````python
# Define the function f(x) = -x^3 - cos(x)
def f(x):
    return -x**3 - np.cos(x)

# Derivative of f(x) for Newton's Method
def f_prime(x):
    return -3*x**2 + np.sin(x)
````

<center><h3>Newton Method</h3></center>

````python
# Newton's Method Implementation
def Newton_Method(p0, tol=0.00001, i_max=100):
    data = []
    p = p0
    
    for i in range(0, i_max):
        fp = f(p)
        f_prime_p = f_prime(p)
        
        # Avoid division by zero
        if f_prime_p == 0:
            print("Derivative is zero, unable to continue with Newton's method.")
            return None

        p_new = p - fp / f_prime_p
        data.append([i, p, fp, f_prime_p, p_new])
        
        # Check if the method converged
        if abs(p_new - p) < tol:
            break
        
        p = p_new

    # Convert the results into a DataFrame
    df = pd.DataFrame(data, columns=["i", "pᵢ", "f(pᵢ)", "f'(pᵢ)", "pᵢ₋₁"])
    return df
````

<center><h3>Secant Method</h3></center>

````python
# Secant Method Implementation
def Secant_Method(p0, p1, tol=0.00001, i_max=100):
    data = []
    for i in range(1, i_max + 1):
        f_p0 = f(p0)
        f_p1 = f(p1)

        # Avoid division by zero
        if f_p1 - f_p0 == 0:
            print("Division by zero in Secant method.")
            return None

        # Secant Method Formula
        p_new = p1 - f_p1 * (p1 - p0) / (f_p1 - f_p0)
        data.append([i, p0, p1, f_p0, f_p1, p_new])

        # Check if the method converged
        if abs(p_new - p1) < tol:
            break

        p0, p1 = p1, p_new

    # Convert the results into a DataFrame
    df = pd.DataFrame(data, columns=["i", "p₀", "p₁", "f(p₀)", "f(p₁)", "pᵢ"])
    return df
````

<center><h3>False Position Method</h3></center>

````python
# False Position Method Implementation (Regula Falsi)
def False_Position_Method(p0, p1, tol=0.00001, i_max=100):
    data = []
    
    # Check initial values
    if f(p0) * f(p1) >= 0:
        print("Invalid initial interval. The function must have different signs at p0 and p1.")
        return None
    
    for i in range(1, i_max + 1):
        f_p0 = f(p0)
        f_p1 = f(p1)

        # False Position Formula
        p_new = p1 - f_p1 * (p1 - p0) / (f_p1 - f_p0)
        f_p_new = f(p_new)

        data.append([i, p0, p1, f_p0, f_p1, p_new, f_p_new])

        # Check if the method converged
        if abs(f_p_new) < tol:
            break

        # Update interval based on the sign of the function
        if f_p0 * f_p_new < 0:
            p1 = p_new
        else:
            p0 = p_new

    # Convert the results into a DataFrame
    df = pd.DataFrame(data, columns=["i", "p₀", "p₁", "f(p₀)", "f(p₁)", "pᵢ", "f(pᵢ)"])
    return df
````
